```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
from scipy.ndimage import minimum_filter, maximum_filter
```

```python
image = cv2.imread('./cat.jpg')

plt.figure(figsize=(5, 5))
plt.imshow(image)
plt.title("Original Image")
plt.axis('off')
plt.show()
```

```python
## Median filter
median_filtered = cv2.medianBlur(image, 3)
plt.figure(figsize=(5, 5))
plt.imshow(median_filtered)
plt.title("Median Filtered Image")
plt.axis('off')
plt.show()
```

```python
# Min filter
plt.figure(figsize=(10, 5))

min_filtered = minimum_filter(image, size=3)

plt.subplot(1, 2, 1)
plt.imshow(min_filtered)
plt.title("Min Filtered Image")
plt.axis('off')

# Max filter
max_filtered = maximum_filter(image, size=3)

plt.subplot(1,2,2)
plt.imshow(max_filtered)
plt.title("Max Filtered Image")
plt.axis('off')
plt.show()
```

```python
rgb_image = cv2.imread('./flower.jpg')

plt.figure(figsize=(5, 5))
plt.imshow(rgb_image)
plt.title("Original Image")
plt.axis('off')
plt.show()
```

```python
plt.figure(figsize=(12, 5))

## Median filter
median_filtered = cv2.medianBlur(rgb_image, 3)

plt.subplot(1, 3, 1)
plt.imshow(median_filtered)
plt.title("Median Filtered Image")
plt.axis('off')

# Min filter
min_filtered = minimum_filter(rgb_image, size=3)
plt.subplot(1,3,2)
plt.imshow(min_filtered)
plt.title("Min Filtered Image")
plt.axis('off')

# Max filter
max_filtered = maximum_filter(rgb_image, size=3)

plt.subplot(1,3,3)
plt.imshow(max_filtered)
plt.title("Max Filtered Image")
plt.axis('off')
plt.show()
```
