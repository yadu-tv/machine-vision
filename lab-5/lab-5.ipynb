Importing libraries

```python
import matplotlib.pyplot as plt # 21BAI1210
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

Loading image

```python

img = cv2.imread('./lenna.png', cv2.IMREAD_GRAYSCALE)
plt.subplot(2,2,1),plt.imshow(img, cmap = 'gray')
plt.title('Original'), plt.xticks([]), plt.yticks([])
plt.show()
```

Sobel Edge Detection

```python
sobelx = cv2.Sobel(img, cv2.CV_64F, 1, 0, ksize=5)
sobely = cv2.Sobel(img, cv2.CV_64F, 0, 1, ksize=5)
sobel = cv2.bitwise_or(sobelx, sobely)
plt.subplot(2,2,2),plt.imshow(sobel, cmap = 'gray')
plt.title('Sobel'), plt.xticks([]), plt.yticks([])
plt.show()
```

Roberts Edge Detection

```python
kernelx = np.array([[-1, 0], [0, 1]], dtype=int)
kernely = np.array([[0, -1], [1, 0]], dtype=int)
robertx = cv2.filter2D(img, -1, kernelx)
roberty = cv2.filter2D(img, -1, kernely)
robert = cv2.bitwise_or(robertx, roberty)
plt.subplot(2,2,3),plt.imshow(robert, cmap = 'gray')
plt.title('Roberts'), plt.xticks([]), plt.yticks([])
plt.show()
```

Prewitt Edge Detection

```python
kernelx = np.array([[-1, -1, -1], [0, 0, 0], [1, 1, 1]], dtype=int)
kernely = np.array([[-1, 0, 1], [-1, 0, 1], [-1, 0, 1]], dtype=int)
prewittx = cv2.filter2D(img, -1, kernelx)
prewitty = cv2.filter2D(img, -1, kernely)
prewitt = cv2.bitwise_or(prewittx, prewitty)
plt.subplot(2,2,4),plt.imshow(prewitt, cmap = 'gray')
plt.title('Prewitt'), plt.xticks([]), plt.yticks([])
plt.show()
```
